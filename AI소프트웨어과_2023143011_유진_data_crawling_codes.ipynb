```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL 설정
base_url = "https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop"
params = {
    "land_id": "0",
    "ausrichtung": "Torwart",
    "spielerposition_id": "1",
    "altersklasse": "o34",
    "jahrgang": "0",
    "kontinent_id": "0",
    "page": 1  # 페이지 번호 추가
}
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
}

# 데이터를 저장할 리스트 생성
players = []

# 여러 페이지 크롤링
while len(players) < 50:
    response = requests.get(base_url, headers=headers, params=params)

    # 상태 코드 확인
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")

        # 테이블 찾기
        table = soup.find("table", class_="items")

        if table:
            rows = table.find_all("tr", class_=["odd", "even"])
            for row in rows:
                if len(players) >= 50:
                    break

                # 선수 이름
                name_tag = row.find("td", class_="hauptlink")
                name = name_tag.a.get_text(strip=True) if name_tag and name_tag.a else "N/A"

                # 소속팀
                team_tag = row.find_all("td", class_="zentriert")[2]
                team = team_tag.img["alt"] if team_tag and team_tag.img else "N/A"

                # 국적
                nationality_tag = row.find("img", class_="flaggenrahmen")
                nationality = nationality_tag["alt"] if nationality_tag else "N/A"

                # 나이
                age_tag = row.find_all("td", class_="zentriert")[1]  # 두 번째 "zentriert" 클래스가 나이 열임
                age = age_tag.get_text(strip=True) if age_tag else "N/A"

                # 시장 가치
                market_value_tag = row.find("td", class_="rechts hauptlink")
                market_value = market_value_tag.get_text(strip=True) if market_value_tag else "N/A"

                # 데이터 저장
                players.append({
                    "Player": name,
                    "Club": team,
                    "Nation": nationality,
                    "Age": age,
                    "Market Value": market_value
                })

        else:
            print("Table not found on the page.")
            break
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        break

    # 다음 페이지로 이동
    params["page"] += 1

# 데이터프레임 생성 및 출력 및 CSV 저장
if players:
    df = pd.DataFrame(players)
    df.index = df.index + 1  # 인덱스가 1부터 시작하도록 조정
    print(df.to_string(index=True))
    # CSV 파일 저장
    df.to_csv("transfermarket_o34_Goalkeeper_top50.csv", index=False, encoding="utf-8-sig")
    print("Data saved to 'transfermarket_o34_Goalkeeper_top50.csv'")
else:
    print("No players found.")
```
